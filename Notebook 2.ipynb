First, you will graph the linear function. 

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt

# Graph the linear function 
a = float(input('Please enter a = '))
b = float(input('Please enter b = '))

f = lambda x: a*x + b # Define the function 
x = np.arange(-10.0,10.0,0.2)
plt.plot(x,f(x),'-')
plt.xlabel('x');plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```python
import numpy as np

data = np.array([[1980, 338.7],
                [1982, 341.2],
                [1984, 344.4],
                [1986, 347.2],
                [1988, 351.5],
                [1990, 354.2],
                [1992, 356.3],
                [1994, 358.6],
                [1996, 362.4],
                [1998, 366.5],
                [2000, 369.4],
                [2002, 373.2],
                [2004, 377.5],
                [2006, 381.9],
                [2008, 385.6],
                [2010, 389.9],
                [2012, 393.8]])
```

```python
import matplotlib.pyplot as plt

plt.plot(data[:,0], data[:,1], 'o')
plt.xlabel('t')                     # Add label to x-axis
plt.ylabel('C (ppm)')                # Add label to y-axis

plt.savefig('f8.png', format='png')

plt.show()
```

We next write a code to graph a linear function. 

```python
from numpy import arange

def graph_linear(m,b):
    f = lambda x: m*x + b
    x = arange(1978, 2014,2)
    plt.plot(x,f(x),'-', data[:,0], data[:,1], 'o')
    plt.xlabel('x');plt.ylabel('y')
    plt.grid(True)
    plt.show()
```

We graph the line passing through the first and last data point. Its equation is given by 
$$
C(t) = 1.722t - 3070.86
$$

```python
graph_linear(m=1.722, b=-3070.86 )
```

The linear regression can be made by using linregress() from the module scipy.stats of the statistical functions

```python
from scipy.stats import linregress

t = data[:,0]
C = data[:,1]
L = linregress(t,C)
m = L.slope
b = L.intercept
print(m)
print(b)
    
```

For more information about linregress(), please [click here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.linregress.html).

We get ready to graph the **regression line**.

```python
graph_linear(m,b)
```

We can test whether our regression line gives a good approximation. 

```python
from pandas import DataFrame
from math import *

regression = lambda x: m*x + b
result1 = {'Year': t, 'Exact value of C': C, 'Approximate value of C': regression(t), 'Error': abs(C-regression(t))}
DataFrame(result1)
```

Moreover, we can estimate the average CO$_2$ level for 1987 and predict the level for the year 2020 as follows. 

```python
print('The average carbon dioxide level for 1987 is ', regression(1987))
print('\nThe average carbon dioxide level for 2020 is ', regression(2020))
```

We can even predict the time when the CO$_2$ exceed 420 parts per million. 

```python
def solveLin(a,b): # solve a linear equation ax+b=0
    return -b/a

m = L.slope
b = L.intercept
t = solveLin(m, b-420)
print('We can predict that the carbon dioxide level will exceed 420 ppm by the year', t)
```

```python
def solveQuad(a,b,c): # solve a quadratic equation
    Delta = (b**2 - 4*a*c)
    x1 = (-b + sqrt(Delta))/(2*a)
    x2 = (-b - sqrt(Delta))/(2*a)
    return x1, x2

# We can find time when the ball hits the ground as follows.
t1, t2 = solveQuad(a = -4.90, b = 0.96, c = 449.36)
print(t1)
print(t2)    
```

## 3.3 Power functions

Graph the function $f(x)=x^a$.

```python
x = np.linspace(-1, 1, endpoint = True)

p1 = lambda x: x
p2 = lambda x: x**2
p3 = lambda x: x**3
p4 = lambda x: x**4
p5 = lambda x: x**5

plt.plot(x, p1(x), label = "$y = x$")
plt.plot(x, p2(x), label = "$y = x^2$")
plt.plot(x, p3(x), label = "$y = x^3$")
plt.plot(x, p4(x), label = "$y = x^4$")
plt.plot(x, p5(x), label = "$y = x^5$")

plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc =0)
plt.show()
```

Next you will learn to graph the root functions.

```python
from math import *

x1 = np.linspace(-1, 0, endpoint = False)
x2 = np.linspace(0, 1, endpoint = True)
x3 = np.concatenate((x1, x2), axis = 0)

r2 = lambda x: x**(1/2)
r3 = lambda x: x**(1/3)

y = np.concatenate((-r3(-x1), r3(x2)), axis = 0)

plt.plot(x, r2(x2), label = "$y = \sqrt{x}$")
plt.plot(x3, y, label = "$y = \sqrt[3]{x}$")

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## 3.7 Exponential Functions

Let us graph the functions $y = 2^x$ and $y=0.5^x$ on the same screen. 

```python
x = np.arange(-3.0, 3.1,0.1)

plt.plot(x, 2**(x), '-', label = "$y = 2^x$")
plt.plot(x, 0.5**(x), '-', label = "$y = 0.5^x$")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
